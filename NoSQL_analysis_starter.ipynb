### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": 20}

# Use count_documents to display the number of documents in the result
count_hygiene_20 = establishments.count_documents(query)
print(f"Number of documents with hygiene score of 20: {count_hygiene_20}")

# Display the first document in the results using pprint
first_hygiene_20 = establishments.find_one(query)
pprint(first_hygiene_20)
```

```python
# Convert the result to a Pandas DataFrame
hygiene_20_df = pd.DataFrame(list(establishments.find(query)))

# Display the number of rows in the DataFrame
print(f"Number of rows in the DataFrame: {hygiene_20_df.shape[0]}")

# Display the first 10 rows of the DataFrame
print(hygiene_20_df.head(10))
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

penang_flavours_location = {"latitude": 51.083812, "longitude": 1.195625}

# Define the range for latitude and longitude within 0.01 degree on either side
lat_min = penang_flavours_location['latitude'] - 0.01
lat_max = penang_flavours_location['latitude'] + 0.01
long_min = penang_flavours_location['longitude'] - 0.01
long_max = penang_flavours_location['longitude'] + 0.01

# Define the query to find establishments with a RatingValue of 5 within the specified lat/long range
query = {
    "RatingValue": 5,
    "geocode.latitude": {"$gte": lat_min, "$lte": lat_max},
    "geocode.longitude": {"$gte": long_min, "$lte": long_max}
}

# Sort by lowest Hygiene score and limit to top 5
top_5_establishments = establishments.find(query).sort("scores.Hygiene", 1).limit(5)



# Print the results
print("Top 5 establishments with RatingValue of 5, sorted by lowest hygiene score, near 'Penang Flavours':")
for establishment in top_5_establishments:
    pprint(establishment)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
# Define the aggregation pipeline
pipeline = [
    {"$match": {"scores.Hygiene": 0}},  # Match documents with a hygiene score of 0
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},  # Group by Local Authority and count
    {"$sort": {"count": -1}},  # Sort by count in descending order
    {"$limit": 10}  # Limit to top 10 results
]

# Execute the aggregation pipeline
results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print(f"Number of documents in the result: {len(results)}")

# Print the first 10 results
print("First 10 results:")
for result in results[:10]:
    pprint(result)
```

```python
# Convert the result to a Pandas DataFrame
results_df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print(f"Number of rows in the DataFrame: {results_df.shape[0]}")

# Display the first 10 rows of the DataFrame
print("\nFirst 10 rows of the DataFrame:")
print(results_df.head(10))
```
